```python
from fuzzywuzzy import process
from fuzzywuzzy import fuzz
import pandas as pd
```

```python
df_account = pd.read_excel("c:\\prashanth\\anaconda\\goodyear.xls") 
```

```python
print(df_account)
```

```python
accountlist = df_account["Account"].tolist()
```

```python
print(accountlist)
```

```python
df_master = pd.read_excel("c:\\prashanth\\anaconda\\master.xls") 
```

```python
print(df_master)
```

```python
masterlist = df_master["MasterAccount"].tolist()
```

```python
finallist = pd.DataFrame(columns=['CCAccount','WLAccount','Score'])
```

```python
print(finallist)
```

```python
i=0
for master in masterlist:
    matchedaccounts = process.extract(master,accountlist,scorer=fuzz.token_set_ratio)
    for matched in matchedaccounts:
        if matched[1] > 89:            
            finallist.loc[i] = [master,matched[0],matched[1]]
            i = i + 1 
```

```python
finallist
```

```python
writer = pd.ExcelWriter('AccountMatching_tokensetration.xlsx',engine='xlsxwriter')
```

```python
finallist.to_excel(writer,sheet_name='Sheet1')
```

```python
writer.save()
```
