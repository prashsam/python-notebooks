```python
import pandas as pd
import numpy as np
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
```

```python
df_settlement = pd.read_excel("C:\\Users\\psamp\\AnacondaProjects\\AsbestosIndemnity.xlsx") 
```

```python
df_settlement
```

```python
df_hotcodingdisease = pd.get_dummies(df_settlement['diseasecategory'])
```

```python
df_hotcodingjurisdiction = pd.get_dummies(df_settlement['Jurisdiction'])
```

```python
df_final = pd.concat([df_settlement,df_hotcodingdisease,df_hotcodingjurisdiction], axis = 1)
```

```python
df_final = df_final.drop(['diseasecategory','Jurisdiction'],axis=1)
```

```python
df_final
```

```python
X = df_final.drop(['indemnitypaid'],axis=1)
y = df_final.indemnitypaid
```

```python
X = X.values
y = y.values
```

```python
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size=0.25,random_state=1 )
```

```python
from sklearn import linear_model

clf = linear_model.SGDRegressor()
clf.fit(X_train, y_train)
```

```python
predicted = clf.predict(X_test)
```
