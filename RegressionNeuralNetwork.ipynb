```python
import pandas as pd
import numpy as np
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dense, Dropout, Activation, Flatten
```

```python
df_settlement = pd.read_excel("C:\\Users\\psamp\\AnacondaProjects\\AsbestosSett.xls") 
```

```python
df_settlement
```

```python
df_hotcodingdisease = pd.get_dummies(df_settlement['diseasecategory'])
```

```python
df_hotcodingjurisdiction = pd.get_dummies(df_settlement['Jurisdiction'])
```

```python
df_final = pd.concat([df_settlement,df_hotcodingdisease,df_hotcodingjurisdiction], axis = 1)
```

```python
df_final = df_final.drop(['diseasecategory','Jurisdiction'],axis=1)
```

```python
df_final
```

```python
X = df_final.drop(['indemnitypaid'],axis=1)
y = df_final.indemnitypaid

X = X.values
y = y.values.reshape((1058,1))
scaler = MinMaxScaler()
y = scaler.fit_transform(y)
```

```python
y
```

```python
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size=0.25,random_state=1 )
#X_train = X_train.values
#X_test = X_test.values
#y_train = y_train.values
#y_test = y_test.values
```

```python
y_train
```

```python
#model=Sequential()
#model.add(Dense(128, input_dim=18, activation='relu'))
#model.add(Dense(32, activation='relu'))
#model.add(Dense(1))

#model=Sequential()
#model.add(Dense(18, input_dim=18, activation='relu'))
#model.add(Dense(9, activation='relu'))
#model.add(Dense(1))

model=Sequential()
model.add(Dense(180, input_dim=18, activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(90, activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(18,activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(loss='mean_squared_error', optimizer='adam')
```

```python
model.fit(X_train, y_train, validation_data = (X_test, y_test), epochs=1000) 
           
```

```python
predicted_value = model.predict(X_test)
```
