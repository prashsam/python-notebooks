```python
import pandas as pd
import numpy as np
from sklearn.multiclass import OneVsRestClassifier
from sklearn.svm import LinearSVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
df_settlement = pd.read_excel("C:\\Users\\psamp\\AnacondaProjects\\AsbestosIndemnity3.xlsx") 
```

```python
df_settlement
```

```python
df_hotcodingdisease = pd.get_dummies(df_settlement['diseasecategory'])
```

```python
df_hotcodingjurisdiction = pd.get_dummies(df_settlement['Jurisdiction'])
```

```python
df_final = pd.concat([df_settlement,df_hotcodingdisease,df_hotcodingjurisdiction], axis = 1)
```

```python
df_final = df_final.drop(['diseasecategory','Jurisdiction'],axis=1)
```

```python
df_final = df_final.drop(['indemnitypaid'],axis=1)
```

```python
df_final
```

```python
X = df_final.drop(['Indemnity_Paid_Bin'], axis = 1)
y = df_final.Indemnity_Paid_Bin

X = X.values
y = y.values
```

```python
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size=0.25,random_state=1 )
```

```python
y_train
```

```python
clf = OneVsRestClassifier(LinearSVC(random_state=0)).fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```
