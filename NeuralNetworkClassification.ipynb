```python
import pandas as pd
import numpy as np
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dense, Dropout, Activation, Flatten
from keras.layers.recurrent import LSTM, GRU, SimpleRNN
```

```python
df_settlement = pd.read_excel("C:\\Users\\psamp\\AnacondaProjects\\AsbestosIndemnity2.xlsx") 
```

```python
df_settlement
```

```python
df_hotcodingdisease = pd.get_dummies(df_settlement['diseasecategory'])
```

```python
df_hotcodingjurisdiction = pd.get_dummies(df_settlement['Jurisdiction'])
```

```python
df_Indemnity_Bin = pd.get_dummies(df_settlement['Indemnity_Paid_Bin'])
```

```python
df_Indemnity_Bin
```

```python
df_final = pd.concat([df_settlement,df_hotcodingdisease,df_hotcodingjurisdiction], axis = 1)
```

```python
df_final = df_final.drop(['diseasecategory','Jurisdiction'],axis=1)
```

```python
df_final = df_final.drop(['indemnitypaid','Indemnity_Paid_Bin'],axis=1)
```

```python
df_final
```

```python
X = df_final
y = df_Indemnity_Bin

X = X.values
y = y.values
```

```python
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size=0.25,random_state=1 )
```

```python
y_train
```

```python

model=Sequential()
model.add(Dense(180, input_dim=18, activation='relu'))
#model.add(Dropout(0.1))
model.add(Dense(90, activation='relu'))
#model.add(Dropout(0.1))
model.add(Dense(45,activation='relu'))
#model.add(Dropout(0.1))
model.add(Dense(21,activation='relu'))
#model.add(Dropout(0.1))
model.add(Dense(7,activation='softmax'))


```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam' , metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, validation_data = (X_test, y_test), epochs=1000) 
           
```
