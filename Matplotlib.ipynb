```python
from matplotlib import pylab
print(pylab.__version__)
```

```python
import numpy as np
```

```python
x = np.linspace(0,10,25)
y = x * x + 2
print(np.array([x,y]).reshape(25,2))
```

```python
pylab.plot(x,y,'r')
```

```python
pylab.subplot(1,2,1)  # row , column , index
pylab.plot(x,y,'r--')
pylab.subplot(1,2,2)  # row , column , index
pylab.plot(y,x,'g*-')
```

```python
pylab.subplot(1,2,1)  # row , column , index
pylab.plot(x,y,'r--')
pylab.subplot(1,2,1)  # row , column , index
pylab.plot(y,x,'g*-')
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
axis = fig.add_axes([0.5,0.1,0.8,0.8])  #control left , right , width , hieght of the canvas (from 0 to 1)
axis.plot(x,y,'r') 
```

```python
fig,axis = plt.subplots(nrows=1,ncols=3)  
for ax in axis:
    ax.plot(x,y,'r' )
```

```python
fig,axis = plt.subplots(nrows=3,ncols=1)  
for ax in axis:
    ax.plot(x,y,'r' )
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])  # big axes
axes2 = fig.add_axes([0.2,0.5,0.3,0.3])  # small axes
axes1.plot(x,y, 'r')
axes2.plot(y,x,'g')
```

```python
fig = plt.figure(figsize=(16,9))
fig.add_subplot()
plt.plot(x,y,'r')
```

```python
fig,axes = plt.subplots()
axes.set_xlabel("x")
axes.set_ylabel("y")
axes.set_title("title")
axes.plot(x,y,'r')
```

```python
fig,axes = plt.subplots()
axes.set_xlabel("x_label")
axes.set_ylabel("y_label")
axes.set_title("graph_title")

axes.plot(x,x**2)
axes.plot(x,x**3)
axes.legend(["y=x**2","y=x**3"],loc=2)
```

```python
fig,axes = plt.subplots(dpi=200)
axes.set_xlabel("x_label")
axes.set_ylabel("y_label")
axes.set_title("graph_title")

axes.plot(x,x**2)
axes.plot(x,x**3)
axes.legend(["y=x**2","y=x**3"],loc=2)
```
